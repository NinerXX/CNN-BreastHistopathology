```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = 'C:/Users/Niner/Desktop/Port/testset'
dimension= 180
batchsize = 10
categories = ['Benign', 'Malignant']
```

```python
data = tf.keras.utils.image_dataset_from_directory(
    path,
    image_size=(dimension, dimension),
    batch_size=batchsize
)
```

```python
model = tf.keras.models.load_model('Breast-Cancer-Diagnosis-CNN.model')
```

```python
plt.figure(figsize=(12, 6))
for image, label in data.take(1):
    for i in range(10):
        plt.subplot(2, 5, i+1)
        plt.imshow(image[i].numpy().astype('uint8'))
        plt.title(f'True:{categories[label.numpy()[i]]}')
        
        prediction = model.predict(np.array(image[i]).reshape(-1, dimension, dimension, 3))
        plt.xlabel(f'Predicted: {categories[np.argmax(prediction[0])]}')
```
